```python
import os
from dotenv import load_dotenv
load_dotenv()

GROQ_API_KEY = os.getenv("GROQ_API_KEY")
assert GROQ_API_KEY
print(f"✅ Key loaded. Starts with: {GROQ_API_KEY[:10]}...  (length: {len(GROQ_API_KEY)})")
```

```python
import requests
url = "https://api.groq.com/openai/v1/chat/completions"

headers = {"Authorization": f"Bearer {GROQ_API_KEY}",
           "Content-Type": "application/json",}

body = {"model": "llama-3.3-70b-versatile",
        "messages": [
            {"role": "system", "content": "You are a helpful Artificial Intelligence tutor. Keep replies under 2 sentences."},
            {"role": "user", "content": "What is a LLM in one line?"},],}

response = requests.post(url, headers = headers, json = body, timeout = 30)
data = response.json()

if "choices" in data:
    print("🦙 Reply from Llama:", data["choices"][0]["message"]["content"])
    print(f"\n💰 Tokens used: {data['usage']['total_tokens']}  (cost: $0.00 — free tier)")
else:
    print("API Error", data)
```

```python
from openai import OpenAI
groq_client = OpenAI(
    api_key = GROQ_API_KEY,
    base_url = "https://api.groq.com/openai/v1",
)

response = groq_client.chat.completions.create(
    model = "llama-3.3-70b-versatile",
    messages = [
         {"role": "system", "content": "You are a helpful math tutor. keep replies under 2 setences "},
         {"role": "user", "content": "What is baye's theorem?"},
     ],
    
)

print("🦙 Reply:", response.choices[0].message.content)
```

```python
models_to_try = [
    "llama-3.3-70b-versatile",
    "llama-3.1-8b-instant",
    "meta-llama/llama-prompt-guard-2-86m",
]

question = "In 1 sentence, what makes Python a great language for beginners?"

for model in models_to_try:
    print(f"\n{'='*60}\n🤖 {model}\n{'='*60}")
    try:
        r = groq_client.chat.completions.create(
            model=model,
            messages=[{"role": "user", "content": question}],
        )
        print(r.choices[0].message.content.strip())
    except Exception as e:
        print(f"⚠️  Failed: {e}")
     
```
